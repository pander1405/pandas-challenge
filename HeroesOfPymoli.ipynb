```python
#Dependencies
import pandas as pd
```

```python
#Take a look at the data file
og_df = pd.read_csv("Rescources/purchase_data.csv")
og_df.sort_values(by='Item ID').head()
```

```python
#View Datatypes
og_df.dtypes 
```

```python
#Check for missing data
og_df.count()

#No missing date, no data clean necessary
```

```python
#Determine player count
unique_players = og_df['SN'].unique()
count_unqiue_players = len(unique_players)
print(f'Total Number of Unique Players: {count_unqiue_players}')
```

```python
#Purchasing Analysis
unique_items = og_df['Item Name'].unique()
count_unique_item =len(unique_items)

av_pur_price = round(og_df['Price'].mean(),2)

total_purchase_count = len(og_df['Purchase ID'])

revenue = round(og_df['Price'].sum(),2)

print(f'Number of Unique Items: {count_unique_item}')
print(f'Average Purchase Pricve: ${av_pur_price}')
print(f'Total Number of Purchases: {total_purchase_count}')
print(f'Total Revenue: ${revenue}')
```

```python
#Gender Demographics
male = og_df.loc[og_df['Gender'] == 'Male']
female = og_df.loc[og_df['Gender'] == 'Female']
other = og_df.loc[og_df['Gender'] == 'Other / Non-Disclosed']

count_male = len(male['SN'].unique())
per_male = round(count_male / len(og_df['SN'].unique())*100,0)

count_female = len(female['SN'].unique())
per_female = round(count_female / len(og_df['SN'].unique())*100,0)

count_other = len(other['SN'].unique())
per_other = round(count_other / len(og_df['SN'].unique())*100,0)

#Another way of doing this. 2 lines of count instead of 6
gender_count = og_df['Gender'].value_counts()

gender_percent = round(og_df['Gender'].value_counts(normalize=True)*100)

print(f'Percent & Count Male Players: {per_male}% & {count_male}')
print(f'Percent & Count Female Players: {per_female}% & {count_female}')
print(f'Percent & Count Other Players: {per_other}% & {count_other}')
```

```python
#Purchase Analysis (Gender)
gped_by_gen = og_df.groupby(['Gender'])

gen_sum = pd.DataFrame({'Purchase Count': gped_by_gen['Purchase ID'].count(),
                       'Average Purchase Count': round(gped_by_gen['Price'].mean(),2),
                       'Total Purchase Value': round(gped_by_gen['Price'].sum(),2),
                       'Average Purchase Total Per Person': round(gped_by_gen['Price'].sum() / gped_by_gen['SN'].nunique(),2)})


gen_sum
```

```python
#Age Demographics


bins = [0,9,14,19,24,29,34,39,50]
g_name =['Less than 10','10-14','15-19','20-24','25-29','30-34','35-39','40 plus']

og_df['Age Summary'] = pd.cut(og_df['Age'], bins, labels=g_name, include_lowest=True)

age_summary_grpd = og_df.groupby('Age Summary')

age_dem = pd.DataFrame({'Purchase Count': age_summary_grpd['Purchase ID'].count(),
                      'Average Purchase Price': round(age_summary_grpd['Price'].mean(),2),
                      'Total Purchase Value': age_summary_grpd['Price'].sum(),
                       'Average Purchase Total per Person': round(age_summary_grpd['Price'].sum() / 
                                                                  age_summary_grpd['SN'].nunique(),2)})

                        
age_dem
```

```python
#Top Spenders
gped_SN = og_df.groupby('SN') 

sum_spend = pd.DataFrame({'Total Spend': gped_SN['Price'].sum(),'Purchase Count' : gped_SN['Purchase ID'].count(),
                          'Av. Pur. Price' : round(gped_SN['Price'].mean(),2),
                          })

top_spenders = sum_spend.sort_values(by='Total Spend', ascending = False).iloc[:5]

top_spenders
```

```python
#Most Popular Items
gped_item = og_df.groupby(['Item ID','Item Name'])

#This is wrong
pur_count = pd.DataFrame({'Purchase Count': gped_item['Purchase ID'].count(),
                          'Item Price': round(gped_item['Price'].sum() / gped_item['Purchase ID'].count(),2), #because price column is purcase ID total no individual item price
                          'Tot Pur Value': gped_item['Price'].sum()})

pur_count.sort_values(by='Item ID').head()

most_popular_items = pur_count.sort_values(by='Purchase Count', ascending = False).iloc[:5]

most_popular_items
```
